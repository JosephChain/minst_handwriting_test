```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print('tensorflow', tf.__version__)
```

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
print('train:', x_train.shape, y_train.shape)
print('test :', x_test.shape, y_test.shape)
print(512*117)
```

```python
# 归一化并扩展通道维度
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0
x_train = np.expand_dims(x_train, -1)  # (N,28,28,1)
x_test  = np.expand_dims(x_test, -1)
num_classes = 10
y_train_cat = keras.utils.to_categorical(y_train, num_classes)
y_test_cat  = keras.utils.to_categorical(y_test, num_classes)
print('after reshape:', x_train.shape, x_test.shape)
```

```python
plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i].squeeze(), cmap='gray')
    plt.title(str(y_train[i]))
    plt.axis('off')
plt.tight_layout()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1
)
datagen.fit(x_train)
# 可视化增强效果（示例）
x_batch = x_train[:1]
it = datagen.flow(x_batch, batch_size=1)
plt.figure(figsize=(6,3))
for i in range(6):
    img = next(it)[0]
    plt.subplot(1,6,i+1)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.axis('off')
plt.suptitle('Augmented examples')
```

```python
def build_model(input_shape=(28,28,1), num_classes=10):
    inputs = keras.Input(shape=input_shape)
    x = layers.Conv2D(32, (3,3), padding='same', activation=None)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(32, (3,3), padding='same', activation=None)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Conv2D(64, (3,3), padding='same', activation=None)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(64, (3,3), padding='same', activation=None)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.25)(x)

    x = layers.Flatten()(x)
    x = layers.Dense(128, activation=None)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = keras.Model(inputs, outputs)
    return model

model = build_model()
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
output_dir = '/mnt/data/mnist_model_checkpoints'
os.makedirs(output_dir, exist_ok=True)
callbacks = [
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6),
    ModelCheckpoint(os.path.join(output_dir, 'best_model.h5'), save_best_only=True, monitor='val_loss')
]
```

```python
batch_size = 256
epochs = 10
steps_per_epoch = x_train.shape[0] // batch_size

history = model.fit(
    datagen.flow(x_train, y_train_cat, batch_size=batch_size),
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    validation_data=(x_test, y_test_cat),
    callbacks=callbacks
)
```

```python
# 假设 model 已经训练好了
output_dir = r"C:\Users\16352\mnist_model_checkpoints"
os.makedirs(output_dir, exist_ok=True)
model.save(os.path.join(output_dir, 'best_model.h5'))
print("模型已保存！")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred_proba = model.predict(x_test)
y_pred = np.argmax(y_pred_proba, axis=1)
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
plt.imshow(cm, interpolation='nearest')
plt.title('Confusion matrix')
plt.colorbar()
plt.xlabel('predicted')
plt.ylabel('true')
plt.show()
```

```python
import os
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image
import matplotlib.pyplot as plt

# -----------------------------
# 1. 加载MNIST模型
# -----------------------------
model_path = r"C:\Users\16352\mnist_model_checkpoints\best_model.h5"
if not os.path.exists(model_path):
    raise FileNotFoundError(f"模型文件不存在: {model_path}")
model = load_model(model_path)
print("模型加载成功！")

# -----------------------------
# 2. 图像预处理函数
# -----------------------------
def preprocess_image(image):
    """灰度化 + 二值化"""
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        gray = image
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    return binary

def resize_to_mnist_format(digit_image):
    """调整单个数字到28x28 MNIST格式"""
    points = np.column_stack(np.where(digit_image > 0))
    if len(points) == 0:
        return np.zeros((28,28,1))
    min_y, min_x = points.min(axis=0)
    max_y, max_x = points.max(axis=0)
    digit_cropped = digit_image[min_y:max_y+1, min_x:max_x+1]
    h, w = digit_cropped.shape
    scale = 20.0 / max(h,w)
    new_h, new_w = int(h*scale), int(w*scale)
    if new_h == 0 or new_w == 0:
        return np.zeros((28,28,1))
    digit_resized = cv2.resize(digit_cropped, (new_w,new_h), interpolation=cv2.INTER_AREA)
    canvas = np.zeros((28,28), dtype=np.uint8)
    y_offset = (28 - new_h)//2
    x_offset = (28 - new_w)//2
    canvas[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = digit_resized
    digit_normalized = canvas.astype('float32')/255.0
    digit_normalized = np.expand_dims(digit_normalized, -1)
    return digit_normalized

def find_and_extract_digits(binary_image):
    """找到数字轮廓并提取每个数字"""
    contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    digits = []
    boxes = []
    # 过滤小轮廓并排序
    filtered = [c for c in contours if cv2.contourArea(c) > 50]
    filtered.sort(key=lambda c: cv2.boundingRect(c)[0])
    for c in filtered:
        x, y, w, h = cv2.boundingRect(c)
        margin = 3
        x, y = max(0,x-margin), max(0,y-margin)
        w = min(binary_image.shape[1]-x, w+2*margin)
        h = min(binary_image.shape[0]-y, h+2*margin)
        roi = binary_image[y:y+h, x:x+w]
        digits.append(resize_to_mnist_format(roi))
        boxes.append((x,y,w,h))
    return digits, boxes

# -----------------------------
# 3. 识别数字函数
# -----------------------------
def recognize_digits(model, digits):
    if len(digits) == 0:
        return np.array([]), np.array([])
    batch = np.array(digits)
    predictions = model.predict(batch, verbose=0)
    recognized = np.argmax(predictions, axis=1)
    confidences = np.max(predictions, axis=1)
    return recognized, confidences

# -----------------------------
# 4. 可视化结果函数
# -----------------------------
def display_results(original, binary, boxes, digits, recognized, confidences):
    """显示原图、二值图、轮廓、识别结果"""
    # 绘制轮廓和识别结果
    contour_image = original.copy()
    if len(contour_image.shape) == 2:
        contour_image = cv2.cvtColor(contour_image, cv2.COLOR_GRAY2RGB)
    for i, (x,y,w,h) in enumerate(boxes):
        cv2.rectangle(contour_image, (x,y), (x+w, y+h), (0,255,0), 2)
        cv2.putText(contour_image, f'{recognized[i]}({confidences[i]:.2f})', (x, y-10),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,0,0), 2)
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.imshow(original if len(original.shape)==2 else cv2.cvtColor(original, cv2.COLOR_BGR2RGB), cmap='gray')
    plt.title('原图')
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(contour_image)
    plt.title('数字检测与识别')
    plt.axis('off')
    plt.show()
    
    # 显示单个数字
    if digits and len(recognized) > 0:
        n = len(recognized)
        cols = min(10, n)
        rows = (n + cols - 1)//cols
        plt.figure(figsize=(15, 2*rows))
        for i in range(n):
            plt.subplot(rows, cols, i+1)
            plt.imshow(digits[i].squeeze(), cmap='gray')
            plt.title(f'{recognized[i]}({confidences[i]:.2f})')
            plt.axis('off')
        plt.tight_layout()
        plt.show()

# -----------------------------
# 5. 主处理函数
# -----------------------------
def process_image_file(file_path):
    if not os.path.exists(file_path):
        print(f"文件不存在: {file_path}")
        return
    print(f"处理文件: {file_path}")
    image = np.array(Image.open(file_path))
    binary = preprocess_image(image)
    digits, boxes = find_and_extract_digits(binary)
    if len(digits) == 0:
        print("未找到数字！")
        return
    recognized, confidences = recognize_digits(model, digits)
    result_str = ''.join(map(str, recognized))
    print(f"\n识别结果: {result_str}")
    print("各个数字置信度:", [f"{c:.2f}" for c in confidences])
    display_results(image, binary, boxes, digits, recognized, confidences)
    return result_str

# -----------------------------
# 6. 使用示例
# -----------------------------
image_path = r"C:\Users\16352\15154.jpg"
process_image_file(image_path)
```
